```python
def logger(func):
    def inner(*args):
        print("引数：",args)
        return func(*args)
    return inner

def accumulate(a,b):
    return a + b

newfunc = logger(accumulate)
print(newfunc(1,2))
```

```python
@logger
def accumulate(a,b):
    return a+b

print(accumulate(1,2))
```

```python
class Prism:
    def __init__(self, width, height, depth):
        self.width = width
        self.height =height
        self.depth = depth
    
    def content(self):
        return self.width * self.height * self.depth

p1 = Prism(10,20,30)
print(p1.content())
print(p1)

p1.depth=40
print(p1.content())
```

```python
class Cube(Prism):
    def __init__(self, length):
        super().__init__(length,length,length)

c=Cube(20)
c.content()
```

```python
class Prop:
    def __init__(self):
        self.__x = 0
    def getx(self):
        return self.__x
    def setx(self,x):
        self.__x = x
    x = property(getx,setx)
    
i = Prop()
print(i.x)

i.x=10
print(i.x)
```

```python
import numpy as np

np.random.seed(9)
x_orig=np.linspace(0,1,100)

def f(x):
    return np.sin(2 * np.pi * x)

x =np.random.uniform(0,1,size=100)[:,np.newaxis]
y = f(x)+np.random.normal(scale=0.3,size=100)[:,np.newaxis]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split#cross_validationは0.20からなくなった

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8)

plt.plot(x_orig,f(x_orig),ls=':')
plt.scatter(x_train,y_train)
plt.xlim((0,1))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

fig,axs = plt.subplots(2,2,figsize=(8,5))

for ax,deg in zip(axs.ravel(),[0,1,3,9]):
    e=make_pipeline(PolynomialFeatures(deg),LinearRegression())
    e.fit(x_train,y_train)
    
    px=e.predict(x_orig[:,np.newaxis])
    ax.scatter(x_train,y_train)
    ax.plot(x_orig,px)
    ax.set(xlim=(0,1),ylim=(-2,2), ylabel='y', xlabel='x', title='degree={}'.format(deg))
    
plt.tight_layout()
```

```python
from sklearn.metrics import mean_squared_error

train_error = np.empty(10)
test_error = np.empty(10)

for deg in range(10):
    e=make_pipeline(PolynomialFeatures(deg),LinearRegression())
    e.fit(x_train,y_train)
    train_error[deg] = mean_squared_error(y_train,e.predict(x_train))
    test_error[deg] = mean_squared_error(y_test,e.predict(x_test))
    
plt.plot(np.arange(10),train_error,ls=':',label='train')
plt.plot(np.arange(10),test_error,ls=':',label='test')
plt.ylim((0,1))
plt.legend(loc='upper left')
```
